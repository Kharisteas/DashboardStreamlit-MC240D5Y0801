## Analisis Lanjutan (Opsional)

```python
#visualisasi untuk pertanyaan 1
plt.figure(figsize=(10,6))
sns.countplot(x='review_score', data=data)
plt.title('Distribusi Skor Ulasan')
plt.xlabel('Skor Ulasan')
plt.ylabel('Jumlah Ulasan')
plt.show()

#menghitung ulasan positif dan negatif
data['sentiment'] = data ['review_score'].apply(lambda x: 'positif' if x>= 4 else 'negatif')

#word cloud untuk pertanyaan 2
positive_reviews = ' '.join(data[data['sentiment'] == 'positif']['review_comment_message'])
negative_reviews = ' '.join(data[data['sentiment'] == 'negatif']['review_comment_message'])

wordcloud_positive = WordCloud(width=800, height=400, background_color='white').generate(positive_reviews)
wordcloud_negative = WordCloud(width=800, height=400, background_color='white').generate(negative_reviews)

plt.figure(figsize=(12, 6))
plt.subplot(1,2,1)
plt.imshow(wordcloud_positive, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud - Ulasan Positif')

plt.subplot(1,2,2)
plt.imshow(wordcloud_negative, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud - Ulasan Negatif')

plt.show()
```

## Conclusion

- Conclution pertanyaan 1 :
  Dari visualisasi distribusi skor ulasan, kita bisa melihat jumlah ulasan yang paling banyak berada pada skor 4 dan 5, yang menunjukkan bahwa sebagian besar pelanggan memberikan ulasan positif. Namun, ada juga ulasan negatif, terutama pada skor 1 dan 2, yang perlu diperhatikan untuk perbaikan kedepannya.

- Conclution pertanyaan 2 : Dari analisis kata-kata yang sering muncul di word cloud, kita dapat melihat bahwa kata-kata positif seperti "bagus", "puas", dan "cepat" lebih sering muncul di ulasan positif. Sebaliknya, dalam ulasan negatif, kata-kata seperti "jelek", "lama", dan "tidak puas" mendominasi, menunjukkan area-area yang perlu diperbaiki
